```
!pip3 install pymagnitude
from pymagnitude import *
import numpy as np

from google.colab import drive
drive.mount('/content/gdrive')

vectors_es = Magnitude("/content/gdrive/My Drive/umwe-esp-smallish.magnitude")

k = 3

nearest_neighbor_average_es = []
index_es = dict()
all_es = []

index = 0
for key, vector in vectors_es:
  neighbors = vectors_es.most_similar(vector, topn = k)
  nearest_neighbor_average_es.append(sum(sim for _, sim in neighbors)/k)

  index_es[index] = key
  all_es.append(vector)
  index += 1
  if index%50 == 0:
    print("Target nearest neighbors word " + str(index) + " of " + str(len(vectors_es)))

vectors_es = None # Free up RAM

vectors_en = Magnitude("/content/gdrive/My Drive/umwe-eng-small.magnitude")

nearest_neighbor_average_es = np.array(nearest_neighbor_average_es).reshape((-1, 1))
all_es = np.array(all_es)
norms_all_es = np.linalg.norm(all_es, axis = 1).reshape((-1, 1))

en_word = []
es_translation = []

count = 0
for key, vector in vectors_en:
  en_word.append(key)
  vector = vector.reshape((300, 1))
  norm = np.linalg.norm(vector)
  scores = 2*np.matmul(all_es,vector)/norm/norms_all_es - nearest_neighbor_average_es
  es_translation.append(index_es[np.argmax(scores)])

  count += 1
  if count%50 == 0:
    print("Translating source word " + str(count) + " of " + str(len(vectors_en)))

# Outputs translations to file with the format english word + tab + best translation
with open("/content/gdrive/My Drive/translations_umwe_eng.txt", "w") as myfile:
    for i in range(len(en_word)):
        myfile.write(en_word[i] + "\t" + es_translation[i] + "\n")
```
